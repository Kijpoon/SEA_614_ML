## Term frequency-inverse document frequency

0. เตรียม vector bag of words สำหรับแปลงเป็น tfidf

4. กำหนดให้ แถวและคอลัมน์ เริ่มที่ 0 
- จงอธิบายค่าที่ได้ ในแถวที่ 2 คอมลัมน์ที่ 7 ซึ่งมีค่าเท่ากับ 1.69
- แสดงการคำนวณค่าดังกล่าว

**ตอบ**
#### อธิบายค่าที่ได้ ในแถวที่ 2 คอมลัมน์ที่ 7 ซึ่งมีค่าเท่ากับ 1.69
ค่า 1.69 ใน row ที่ 2 columns ที่ 7 ใช้บอกโมเดลว่าคำว่า "two" เป็นคำสำคัญ (Keyword) ที่ใช้เป็นเอกลักษณ์เพื่อแยกแยะประโยคที่ 3 ออกจากประโยคอื่น ๆ ได้เป็นอย่างดี เนื่องจากคำว่า two โผล่มาแค่ในประโยคนี้ประโยคเดียว

#### แสดงการคำนวณค่าดังกล่าว
ขั้นตอนที่ 1: หาค่า TF (Term Frequency)
ดูว่าคำว่า two โผล่มาในประโยคนี้กี่ครั้ง
คำว่า "two" ปรากฏในประโยคที่ 3 จำนวน $1$ ครั้ง -> ค่า TF เท่ากับ 1
#### -------------------------------------------------------------------------------------------------------------------------------------------------------------------------
ขั้นตอนที่ 2: ประเมินว่าคำนี้หายากแค่ไหนจากประโยคทั้งหมดจากสูตร:
$$IDF = \ln(\frac{N + 1}{df + 1}) + 1$$

$N$ (จำนวนประโยคทั้งหมด) = $3$ ประโยค และ $df$ (Document Frequency หรือ จำนวนประโยคที่มีคำว่า "two" ซ่อนอยู่) = $1$ ประโยค (เจอแค่ในประโยคที่ 3 เท่านั้น)

แทนค่าลงในสมการ:
$$IDF = \ln(\frac{3 + 1}{1 + 1}) + 1$$$$IDF = \ln(\frac{4}{2}) + 1$$$$IDF = \ln(2) + 1$$$$IDF \approx 0.693 + 1 = 1.693$$

#### -------------------------------------------------------------------------------------------------------------------------------------------------------------------------
ขั้นตอนที่ 3: นำค่าที่ได้มาคูณกัน
$$TF\text{-}IDF = TF \times IDF$$$$TF\text{-}IDF = 1 \times 1.693 = 1.693$$

## Cleaning text data

0. สร้างฟังก์ชันสำหรับทำความสะอาดข้อความ

```python
import re
def preprocess(text):
    text = re.sub('<[^>]*>', '', text)
    emoticons = re.findall(r'(?::|;|=)(?:-)?(?:\)|\(|D|P)',text)
    text = (re.sub(r'[\W]+', ' ', text.lower()) +
            ' '.join(emoticons).replace('-', ''))
    return text
```

## สร้างโมเดล logistic regression model สำหรับจำแนกข้อความ  

1. อ่านไฟล์ movie_data.csv

```python
import pandas as pd

df = pd.read_csv('movie_data.csv', encoding='utf-8')
df.head(3)
```

2. สร้างคำหยุด จากคำหยุดมาตรฐานในภาษาอังกฤษ 'english' เก็บไว้ในตัวแปร stop

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
```

3. สร้างฟังก์ทำความสะอาดข้อความ ชื่อฟังก์ชัน TextClaen

```python
import re
def TextClean (text):
    text = re.sub('<[^>]*>', '', text)
    emoticons = re.findall(r'(?::|;|=)(?:-)?(?:\)|\(|D|P)',
                           text)
    text = (re.sub(r'[\W]+', ' ', text.lower()) +
            ' '.join(emoticons).replace('-', ''))
    return text
```

4. แบ่งชุดข้อมูล train และ test โดย train คือ  25000 ข้อความแรก และ test คือ  25000 ข้อความหลัง 

```python
X_train = df.loc[:25000]['review'].values
y_train = df.loc[:25000]['sentiment'].values
X_test = df.loc[25000:]['review'].values
y_test = df.loc[25000:]['sentiment'].values
```

5. สร้างรูปแบบตัวสกัดคุณลักษณะ ด้วย TfidfVectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(use_idf=True, norm= 'l2', smooth_idf=True,stop_words=stop, preprocessor=TextClean)
```

6. fit และ transform ตัวสกัดคุณลักษณะบนข้อมูล X_train เก็บไว้ที่ตัวแปร V_train

```python
V_train = tfidf.fit_transform(X_train)
```

7. transform ตัวสกัดคุณลักษณะบนข้อมูล X_test เก็บไว้ที่ตัวแปร V_test

```python
V_test = tfidf.transform(X_test)
```

8. สร้างโมเดลจำแนกโดยใช้ LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0, solver='liblinear')
```

9. สอนโมเดลด้วยชุดข้อมูลสอน V_train,y_train

```python
clf.fit(V_train,y_train)
```

10. ทำนาย ด้วยชุดข้อมูลทดสอบ V_test

```python
y_pred = clf.predict(V_test)
```

11. หาค่า confusion matrix , accuracy , precision, recall, f1

```python
from sklearn.metrics import classification_report, confusion_matrix
cfm = confusion_matrix(y_test,y_pred)
summary_eval = classification_report(y_test,y_pred)
print(cfm)
print(summary_eval)
```

## Review Shopping

สร้างโมเดลโดยใช้ข้อมูลจาก review_shopping.csv และ Deploy model บน Streamlit

โดยสร้างอย่างน้อย 2 โมเดล และเลือกที่ดีกว่ามา Deploy

```python
def thai_tokenize(text):
    return tokenize(text)
```

```python
rv_data = pd.read_csv('review_shopping.csv', sep='\t', header=None, names=['text', 'sentiment'])
rv_data.head()
```

```python
vectorizer = TfidfVectorizer(tokenizer=thai_tokenize, ngram_range=(1, 2))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(rv_data['text'], rv_data['sentiment'], test_size=0.2, random_state=42)

X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.naive_bayes import MultinomialNB

rv_lr = LogisticRegression(random_state=42, solver='liblinear')
rv_lr.fit(X_train_vec, y_train)

y_pred = rv_lr.predict(X_test_vec)
```

```python
score_lr = rv_lr.score(X_test_vec, y_test)

rv_cfm = confusion_matrix(y_test, y_pred)
rv_sum = classification_report(y_test, y_pred)

print(rv_cfm)
print(rv_sum)
print(f"✅ Accuracy (with Attacut): {score_lr:.4f}")
```

```python
multi_nb = MultinomialNB()
multi_nb.fit(X_train_vec, y_train)
y_pred = multi_nb.predict(X_test_vec)
```

```python
mnb_score = multi_nb.score(X_test_vec, y_test)

mnb_cfm = confusion_matrix(y_test, y_pred)
mnb_sum = classification_report(y_test, y_pred)

print(mnb_cfm)
print(mnb_sum)
print(f"✅ Accuracy (with Attacut): {mnb_score:.4f}")
```
